```python
import pandas as pd
```

```python
titanic = pd.read_csv('train.csv')
t = titanic.set_index(['PassengerId'])
```

```python
new = t['Ticket'].unique()
```

توزیع تیکتهای مختلف بر اساس تماما عددی بودن یا شامل عدد و حرف

```python
round(t['Ticket'].str.isnumeric().value_counts(normalize=True)*100, 2)
```

تمامی تیکت های شامل حرف و عدد

```python
ticket_str = t[~(t['Ticket'].str.isnumeric())]['Ticket']
ticket_str.unique()
```

تمامی تیکتهای صرفا عددی

```python
ticket_digit = t[t['Ticket'].str.isnumeric()]['Ticket']
ticket_digit.unique()
```

وضعیت بلیطهای تکراری و غیر تکراری

```python
round(t['Ticket'].duplicated(keep=False).value_counts(normalize=True)*100, 2)
```

تمامی بلیطهایی که بیش از یکبار تکرار شدند شامل هم عددی ها و هم رشته ها

```python
#method 1
ff = t['Ticket'].value_counts()[t['Ticket'].value_counts() > 1]
#method 2
ff = t[t['Ticket'].duplicated(keep=False)]['Ticket'].value_counts()

ff[ff > 1], len(ff[ff > 1])
```

تمامی بلیطهایی که بیش از یکبار تکرار شدند شامل رشته‌ای ها

```python
ff  = ticket_str[ticket_str.duplicated(keep=False)].value_counts()
ff[ff > 1], len(ff[ff > 1])
```

تمامی بلیطهایی که بیش از یکبار تکرار شدند فقط عددی ها

```python
ff  = ticket_digit[ticket_digit.duplicated(keep=False)].value_counts()
ff[ff > 1], len(ff[ff > 1])
```

```python
ff = t[t['Ticket'].duplicated(keep=False)]['Ticket'].value_counts()

for i in ff.items():
    print(f'Ticket: {i[0]}, duplicate in {i[1]} times')
    for column in t.columns:
        if column in ['Name', 'Ticket']:
            continue
        percent = round(t[t['Ticket'] == i[0]].fillna(-1)[column].value_counts(normalize=True)*100, 2).max()
        column_value = t[t['Ticket'] == i[0]].fillna(-1)[column].value_counts().sort_values(ascending=False).head(1).index[0]
        print(f'\t\t\t{percent}% tell {column} is {column_value}')
```

```python
ff = t[t['Ticket'].duplicated(keep=False)]['Ticket'].value_counts()

for column in t.columns:
    if column in ['Name', 'Ticket']:
        continue
    for i in ff.items():
        percent = round(t[t['Ticket'] == i[0]].fillna(-1)[column].value_counts(normalize=True)*100, 2).max()
        column_value = t[t['Ticket'] == i[0]].fillna(-1)[column].value_counts().sort_values(ascending=False).head(1).index[0]
        print(f'Ticket: {i[0]}, duplicate in {i[1]} times and {percent}% tell {column} is {column_value}')
    print("----------------------------------------------------------")   
```

```python
ff = t[t['Ticket'].duplicated(keep=False)]['Ticket'].value_counts()

for column in t.columns:
    sum = 0
    if column in ['Name', 'Ticket']:
        continue
    for i in ff.items():
        percent = round(t[t['Ticket'] == i[0]].fillna(-1)[column].value_counts(normalize=True)*100, 2).max()
        column_value = t[t['Ticket'] == i[0]].fillna(-1)[column].value_counts().sort_values(ascending=False).head(1).index[0]
        sum += i[1] * percent

    print(f'{column}:    \t', sum/ff.sum())
    print("----------------------------------------------------------")   
```

مقادیر زیر درصد مشابهت ردیف ها با تیکت های تکراری در دیگر ستونها را نوشته است.
به عبارتی اگر برای مثال با تیکت های مشابه، 
به این معنی که اگر داده ای که وارد میشود ، شماره‌ی تیکت تکراری داشته باشد، با احتمال های نوشته شده در زیر مقدار ستونهای مختلف آن، مشابه با تیکت تکراری آن در داده ها میباشد.
