```python
import pandas as pd
```

```python
titanic = pd.read_csv('train.csv')
t = titanic.set_index(['PassengerId'])
pd.set_option('display.max_rows', None)
```

```python
result = pd.DataFrame(index=t.columns, columns=t.columns)
```

```python
for outer_column in t.columns:
    ff = t[t[outer_column].duplicated(keep=False)][outer_column].value_counts()
    if outer_column in ['Name']:
        continue
    for column in t.columns:
        sum = 0
        if column in ['Name', outer_column]:
            continue
        for i in ff.items():
            percent = round(t[t[outer_column] == i[0]].fillna(-1)[column].value_counts(normalize=True)*100, 2).max()
            column_value = t[t[outer_column] == i[0]].fillna(-1)[column].value_counts().sort_values(ascending=False).head(1).index[0]
            sum += i[1] * percent
        result[outer_column][column] = round(sum / ff.sum(), 2)

result.dropna(axis='index', how='all').dropna(axis='columns', how='all')
```

ایده این بود:
بر اساس هر ستون آمدم، آنهایی که تکرارشون بیش از یکبار هست را پیدا کردم
خروجی میشد لیستی از اون ستون که تعداد تکرارش از یکبار بیشتر هست.
مثلا بر اساس ستون تیکت اومدم اونهایی که تکرارشون بیش از یکبار هست را در آوردم:

```python
ff = t[t['Ticket'].duplicated(keep=False)]['Ticket'].value_counts()
ff
```

در گام بعدی گفتم من میخوام از بین اونهایی که تکراری هستند 
برای مثال در اینجا تیکت 1601 که اومده و هفت بار تکرار شده، من میخوام 
توزیع این هفت عنصر را روی تمامی ستونهای دیگه نگاه کنم.
(یعنی حالا که هفت تا تیکت تکراری دارم، وضعیت این هفت تا تیکت تکراری تو ستونهای دیگه نسبت به هم چطوری هست)

```python
print(f'Total Ticket {ff.index[0]} is {ff.values[0]}')

for column in t.columns:
    if column in ['Name', 'Ticket']:
        continue
    for i in ff.items():
        percent = round(t[t['Ticket'] == i[0]].fillna(-1)[column].value_counts(normalize=True)*100, 2)#.max()
        column_value = t[t['Ticket'] == i[0]].fillna(-1)[column].value_counts().sort_values(ascending=False)#.head(1).index[0]
        print(column_value, percent)
        print("===============================")
        break
```

از مقادیری که در بالا هست من بیشترین درصد را به همراه لیبلی که بیشترین توزیع را ثبت کرده به عنوان نتیجه‌ی نهایی خارج میکنم
برای مثال برای تیکت 1601 وقتی توزیع روی سروایود را نگاه کنیم، 
نزدیک به هفتاد درصد از تیکتهایی که  شماره 1601 را دارند، مقدار سروایود آنها یک بوده است
پس من بالاترین درصد گزارش شده و مقدار لیبل آن که اینجا یک هست را به عنوان توزیع سروایود ها روی تیکت ۱۶۰۱ در نظر میگیرم.

همین کار را برای تمامی تیکت های دیگر که بیش از یکبار تکرار  شده اند انجام خواهم داد.

توجه شود که نه تنها توزیع سروایود بلکه توزیع روی تمامی ستونها را بررسی میکنم که در زیر آمده است.

```python
ff = t[t['Ticket'].duplicated(keep=False)]['Ticket'].value_counts()

for column in t.columns:
    if column in ['Name', 'Ticket']:
        continue
    for i in ff.items():
        percent = round(t[t['Ticket'] == i[0]].fillna(-1)[column].value_counts(normalize=True)*100, 2).max()
        column_value = t[t['Ticket'] == i[0]].fillna(-1)[column].value_counts().sort_values(ascending=False).head(1).index[0]
        print(f'Ticket: {i[0]}, duplicate in {i[1]} times and {percent}% tell {column} is {column_value}')
    print("----------------------------------------------------------")   
```

آخرین کاری که انجام میدهم، این هست که میانگین وزن دار توزیع ها را حساب میکنم.
در واقع براساس تیکت هایی که بیش از یکبار تکرار شده بودند
من توزیع این تیکت ها را روی تمامی ستونها بدست آوردم

ولی یک سوال اساسی باید برای من حل شود و آن این است که :
بالاخره برای مثال توزیع ستون سروایود روی تیکت چگونه هست؟

برای این کار انتخاب من میانگین وزندار هست
یعنی من برای مثال فهمیدم که تیکت 1601 هفت بار تکرار شده و توزیع آن هم روی سرواید حدود هفتاد درصد میباشد.
همین را برای تیکت
'CA. 2343'
هم فهمیدم که آن هم مثلا هفت بار تکرار شده و توزیع آن صد درصد هست، یعنی همشون سروایود یکسانی دارند 

من در نهایت برای اینکه بالاخره بگم توزیع نهایی چطوری هست، از میانگین وزندار استفاده میکنم
و بین تمامی تیکت هایی که بیش از یکبار تکرار شدند، بر اساس تعداد تکرارشون و بالاترین درصدی که گزارش کردند، میانگین وزندار میگیرم،

و در نهایت میتوانم بگویم میانگین وزندار توزیع سروایودها روی تیکتهای تکراری به چه صورت هست.
و این کار را برای تمامی ستونهای دیگه به غیر از سروایود هم انجام میدم
و در واقع میانگین وزندار توزیع تمامی ستونها را روی تیکت های تکراری خواهم دید

```python
ff = t[t['Ticket'].duplicated(keep=False)]['Ticket'].value_counts()

for column in t.columns:
    sum = 0
    if column in ['Name', 'Ticket']:
        continue
    for i in ff.items():
        percent = round(t[t['Ticket'] == i[0]].fillna(-1)[column].value_counts(normalize=True)*100, 2).max()
        column_value = t[t['Ticket'] == i[0]].fillna(-1)[column].value_counts().sort_values(ascending=False).head(1).index[0]
        sum += i[1] * percent

    print(f'{column}:    \t', sum/ff.sum())
    print("----------------------------------------------------------")   
```

همین روند فوق که برای ستون تیکت رفتم را برای تمامی ستونهای دیگر هم میرم در واقع جدولی که در ادامه آمده است نتیجه‌ی همان هست
که در این جا ستونهایش به ترتیب نقش همین تیکت که تاکنون صحبت شد را بازی میکنند.
یعنی در هر ستون، من به دنبال آنهایی رفتم که تعداد تکرارشان بیش از یکبار هست 
و سپس میانگین وزندار توزیع تمامی ردیفهای این جدول را روی این ستون گفته شده زده یادداشت کرده ام
اگر ستون تیکت را مشاهده کنید، دقیقا خروجی بخش بالا را مشاهده خواهید کرد:

```python
for outer_column in t.columns:
    ff = t[t[outer_column].duplicated(keep=False)][outer_column].value_counts()
    if outer_column in ['Name']:
        continue
    for column in t.columns:
        sum = 0
        if column in ['Name', outer_column]:
            continue
        for i in ff.items():
            percent = round(t[t[outer_column] == i[0]].fillna(-1)[column].value_counts(normalize=True)*100, 2).max()
            column_value = t[t[outer_column] == i[0]].fillna(-1)[column].value_counts().sort_values(ascending=False).head(1).index[0]
            sum += i[1] * percent
        result[outer_column][column] = round(sum / ff.sum(), 2)

result.dropna(axis='index', how='all').dropna(axis='columns', how='all')
```
