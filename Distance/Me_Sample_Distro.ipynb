```python
import pandas as pd
import numpy as np
import sklearn as sk
import math
```

```python
dist = pd.read_csv('dist_data.csv', sep=';')
```

```python
dist_clean = dist[dist['dist'] != "[[], []]"]
```

```python
x = dist_clean.iloc[:, :-1].values
y = dist_clean.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30)
```

```python
def clean_sample(sample):
    sample_list = sample[0].replace(', [', '; [').split(';')
    sample_x = sample_list[0]
    sample_Px = sample_list[1]

    sample_x_list = []
    sample_Px_list = []
    
    for x in sample_x.split(','):
        sample_x_list.append(x.replace(' ', '').replace('[', '').replace(']', ''))
    for Px in sample_Px.split(','):
        sample_Px_list.append(Px.replace(' ', '').replace('[', '').replace(']', ''))

    dict_sample = {}
    for i in range(len(sample_x_list)):
        dict_sample[sample_x_list[i]] = sample_Px_list[i]
    return dict_sample
```

```python
y_pred = []
for dd, iter in enumerate(x_test):
    # if dd != 6:
    #     continue
    best_labels = []
    target = clean_sample(iter)
    for target_key, target_value in target.items():
        Prob = []
        for i in range(len(x_train)):
            tmp = clean_sample(x_train[i]).get(target_key)
            if tmp is not None:
                Prob.append({y_train[i]: abs(float(target_value) - float(tmp))})

        
        #print(Prob)
        min = 9999
        for sample in Prob:
            for key, value in sample.items():
                if value < min:
                    best_label = key
                    min = value

        
        best_labels.append({target_key:(best_label, target_value)})

    #print(best_labels)
    #print(best_labels)
    
    dict_values = {}
    for i in best_labels:
        for value in i.values():
            if dict_values.get(value[0]) is None:
                dict_values[value[0]] = float(value[1])
            else:
                dict_values[value[0]] = dict_values[value[0]] + float(value[1])

    #print(dict_values)
    max = 0
    final_label = 0
    for key, value in dict_values.items():
        if value > max:
            max = value
            final_label = key

    #print(final_label)
    y_pred.append(final_label)    
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
res1 = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(res1)
res11 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (res11)
res12 = accuracy_score(y_test,y_pred)
print("Accuracy:",res12)
```
