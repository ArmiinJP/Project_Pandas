```python
import pandas as pd
```

```python
titanic = pd.read_csv('train.csv')
t = titanic.set_index(['PassengerId'])
```

```python
new = t['Ticket'].unique()
```

توزیع تیکتهای مختلف بر اساس تماما عددی بودن یا شامل عدد و حرف

```python
round(t['Ticket'].str.isnumeric().value_counts(normalize=True)*100, 2)
```

تمامی تیکت های شامل حرف و عدد

```python
ticket_str = t[~(t['Ticket'].str.isnumeric())]['Ticket']
ticket_str.unique()
```

تمامی تیکتهای صرفا عددی

```python
ticket_digit = t[t['Ticket'].str.isnumeric()]['Ticket']
ticket_digit.unique()
```

```python
وضعیت بلیطهای تکراری و غیر تکراری
```

```python
round(t['Ticket'].duplicated().value_counts(normalize=True)*100, 2)
```

تمامی بلیطهایی که بیش از یکبار تکرار شدند شامل هم عددی ها و هم رشته ها

```python
ff = t[t['Ticket'].duplicated()]['Ticket'].value_counts()
ff[ff > 1], len(ff[ff > 1])
```

```python
تمامی بلیطهایی که بیش از یکبار تکرار شدند شامل رشته‌ای ها
```

```python
ff  = ticket_str[ticket_str.duplicated()].value_counts()
ff[ff > 1], len(ff[ff > 1])
```

```python
تمامی بلیطهایی که بیش از یکبار تکرار شدند فقط عددی ها
```

```python
ff  = ticket_digit[ticket_digit.duplicated()].value_counts()
ff[ff > 1], len(ff[ff > 1])
```
