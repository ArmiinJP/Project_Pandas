```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('train.csv')
t = titanic.set_index(['PassengerId'])
sns.set_style('whitegrid')
```

```python
for column in ['Survived', 'Sex', 'Embarked', 'Pclass']:
    labels = []
    sizes = []
    for value in t[column].drop_duplicates().dropna():
        labels.append(value)
        sizes.append(t[t[column] == value].shape[0])
    plt.figure(figsize=(3, 3))
    plt.title(column)
    plt.pie(sizes, labels=labels, autopct='%1.1f%%')
    plt.show()
```

```python
total_count = len(t)
columns = ['SibSp', 'Parch']

for column in columns:
    result = {}
    list_print = []
    sns.catplot(x=column, kind='count', data=t, height=4)
    value_count = t[column].value_counts()
    
    for index, value in value_count.items():
        result[index] = (value / total_count) * 100
    
    for k, v in result.items():
        list_print.append('{}\n{:.1f}%'.format(k, v))

    plt.gca().set_xticklabels(list_print)
    plt.show()
```
