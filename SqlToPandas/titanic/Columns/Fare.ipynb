```python
import pandas as pd
```

```python
titanic = pd.read_csv('train.csv')
t = titanic.set_index(['PassengerId'])
pd.set_option('display.max_rows', None)
```

```python
ff = t[t['Fare'].duplicated(keep=False)]['Fare'].value_counts()

for i in ff.items():
    print(f'Fare: {i[0]}, duplicate in {i[1]} times')
    for column in t.columns:
        if column in ['Name', 'Fare']:
            continue
        percent = round(t[t['Fare'] == i[0]].fillna(-1)[column].value_counts(normalize=True)*100, 2).max()
        column_value = t[t['Fare'] == i[0]].fillna(-1)[column].value_counts().sort_values(ascending=False).head(1).index[0]
        print(f'\t\t\t{percent}% tell {column} is {column_value}')
```
