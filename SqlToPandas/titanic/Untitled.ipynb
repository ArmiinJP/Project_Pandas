```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
```

```python
titanic = pd.read_csv('train.csv')
tips = titanic.set_index(['PassengerId'])
tips.nunique()
```

```python
new1 = tips.groupby(by=['SibSp', 'Parch', 'Sex', 'Survived', 'Pclass']).size().unstack()
new1
```

```python
sns.relplot(
    data=new1,
    x="count", y="SibSp", col="Survived",
    hue="Survived", style="Survived", size='Survived',
)
matplotlib.pyplot.show()
```
