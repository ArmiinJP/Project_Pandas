```python
import numpy as np  
from scipy.stats import norm  
from matplotlib import pyplot as plt
# creating the data distribution
x = np.arange(-5, 5, 1)  
p = norm.pdf(x, 0, 2)  
q = norm.pdf(x, 2, 4)
#compute KL Divergence
"""KL Divergence(P|Q)"""
def KL_div(p_probs, q_probs):    
    KL_div = p_probs * np.log(p_probs / q_probs)
    return np.sum(KL_div)
```

```python
plt.title('KL(P||Q) = %1.3f' % KL_div(p, q))  
plt.plot(x, p)
plt.plot(x, q, c='red')
```

```python
plt.title('KL(Q||P) = %1.3f' % KL_div(q, p))  
plt.plot(x, p)  
plt.plot(x, q, c='red')
```

```python
from scipy import stats
from scipy.stats import norm
import numpy as np
# create the data distribution
data_1 = abs(np.random.randn(1000))
data_2 = np.random.lognormal(size=1000)
#compute KL Divergence
"""KL Divergence(P|Q)"""
def KL_div(p_probs, q_probs):    
    KL_div = p_probs * np.log(p_probs / q_probs)
    return np.sum(KL_div)
def JS_Div(p, q):
    p = np.asarray(p)
    q = np.asarray(q)
    # normalize
    p /= p.sum()
    q /= q.sum()
    m = (p + q) / 2
    return (KL_div(p, m) + KL_div(q, m)) / 2
# JS Divergence is symmetric
result_JSD12= JS_Div(data_1, data_2)
print("JS Divergence between data_1 and data_2",result_JSD12)
result_JSD21= JS_Div(data_2, data_1)
print("JS Divergence between data_2 and data_1",result_JSD21)
```

```python
from scipy.stats import ks_2samp
import numpy as np  
from scipy.stats import norm  
from matplotlib import pyplot as plt
x = np.arange(-5, 5, 1)  
p = norm.pdf(x, 0, 2)  
q = norm.pdf(x, 2, 2)
#perform Kolmogorov-Smirnov test
result=ks_2samp(p, q)
#Plot the data distribution and the KS test results
plt.title( result)  
plt.plot(x, p)  
plt.plot(x, q, c='red')
```
